```python
import time
import openpyxl
import pandas as pd
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

#taking file name as user input
file_name = input('Enter File Name: ')                        # enter the file name with .csv extension (eg - file.csv)

def opentracker():

    def details(i):
        input = driver.find_element('xpath','//input[@name="ip"]')
        input.clear()
        input.send_keys(i)
        input.send_keys(Keys.RETURN)
        time.sleep(3)

        dets = {}
        dets['IP'] = i
        dets['City'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[2]').text.split(': ')[1]
        dets['Region'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[3]').text.split(': ')[1]
        dets['Country'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[4]').text.split(': ')[1]
        dets['Timezone'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[7]').text.split(': ')[1]
        dets['Country code'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[9]').text.split(': ')[1]
        dets['Latitude'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[11]').text.split(': ')[1]
        dets['Longitude'] = driver.find_element('xpath','//*[@id="search_ip_pix1"]/ul/li[12]').text.split(': ')[1]
        dets['ISP'] = driver.find_element('xpath','//*[@id="search_ip_pix2"]/ul/li[4]').text.split(': ')[1]
        dets['Connection'] = driver.find_element('xpath','//*[@id="search_ip_pix2"]/ul/li[5]').text.split(': ')[1]
        return dets

    inp = pd.read_csv(file_name)

    opt = Options()
    opt.add_argument("--incognito")
    global driver
    driver = uc.Chrome(options=opt)
    driver.maximize_window()
    driver.get("https://www.opentracker.net/feature/ip-tracker?ip=")
    time.sleep(1)

    global df_main_opentracker
    df_main_opentracker = pd.DataFrame()
    for i in inp['IP']:
        df = pd.DataFrame()
        dets = details(i)
        df = pd.DataFrame.from_dict(dets,orient='index').transpose()
        df_main_opentracker = pd.concat([df_main_opentracker,df],ignore_index = True)
    driver.quit()
    df_main_opentracker['Source'] = 'Open Tracker'

def ip2location():

    def information(i):
        time.sleep(3)
        enter_ip = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[1]/div[2]/form/div/input')
        enter_ip.clear()
        enter_ip.click()
        enter_ip.send_keys(i)
        enter_ip.send_keys(Keys.RETURN)
        time.sleep(3)

        info = {}
        info['IP'] = i
        info['City'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[5]/td').text
        info['Region'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[4]/td').text
        info['Country'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[3]/td/a').text
        info['Timezone'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[29]/td').text
        info['Country code'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[3]/td').text.split(' [')[1].split(']')[0]
        info['Latitude'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[6]/td').text.split(',')[0]
        info['Longitude'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[6]/td').text.split(', ')[-1].split(' (')[0]
        info['ISP'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[7]/td').text
        info['Connection'] = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div[2]/div[2]/div[3]/table/tbody/tr[10]/td').text
        return info
    
    inp = pd.read_csv(file_name)

    opt = Options()
    opt.add_argument("--incognito")
    global driver
    driver = uc.Chrome(options=opt)
    driver.maximize_window()
    driver.get("https://www.ip2location.com/demo")

    global df_main_ip2location
    df_main_ip2location = pd.DataFrame()
    for i in inp['IP']:
        df = pd.DataFrame()
        time.sleep(5)
        info = information(i)
        df = pd.DataFrame.from_dict(info,orient='index').transpose()
        df_main_ip2location = pd.concat([df_main_ip2location,df],ignore_index = True)
    driver.quit()
    df_main_ip2location['Source'] = 'IP2 Location'

def combining_df():
    global combined_dataframe
    combined_dataframe = pd.concat([df_main_opentracker, df_main_ip2location], ignore_index=True)
    combined_dataframe = combined_dataframe.reset_index(drop=True)
    combined_dataframe.to_csv("IP DATA.csv",index=False)
    print('All information on the given IP addresses are extracted and saved within the same location')

opentracker()
ip2location()
combining_df()
```
